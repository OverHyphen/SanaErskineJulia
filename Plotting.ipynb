#### Basic Plotting:

The most basic plots are line plots.
Here is an example below:

```julia
x = 1:10 #x values that range from 1-10
y = rand(10) # Creates random y values(0-1)

plot(x, y)
```

Anoter Way we can plot data is by giving direct values:

```julia
#plot(x-values, y-values)
plot([0,1,2,3,4,5], [0,2,4,6,8,10])
```

#### Plot Atributes:

The modifiers to plots are called attributes. We can do thing like give the plot a name, or change the line width. 
- Title
- Line Width
- Lable Names
- X-Axis Lable
- Y-Axis Lable
- Plot Type

```julia
x = 1:10
y = rand(10)
plot(x, y, title = "Plot Title", xlabel = ["xAxis"],ylabel = ["yAxis"],label = ["Line 1"], lw = 5)
```

#### Plot Series:

Other ways of plotting a series is called a series type. A line is one series type. However, a scatter plot is another series type which is commonly used. We can change the series type by the seriestype attribute:

```julia
plot(x, y, seriestype = :scatter, title = "My Scatter Plot")
```

Another way to make the Plot

```julia
scatter(x, y, title = "My Scatter Plot")
```

#### Mutiple Plots:

We can combine multiple plots together as subplots using layouts.

```julia
y = rand(10, 4)
plot(x, y, layout = (4, 1))
```

We can also include multiple plots like this:

```julia
p1 = plot(x, y) # Make a line plot
p2 = scatter(x, y) # Make a scatter plot
p3 = plot(x, y, xlabel = "This one is labelled", lw = 3, title = "Subtitle")
p4 = histogram(x, y) # Four histograms each with 10 points? Why not!
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```
